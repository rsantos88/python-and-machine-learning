````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,10)
y = x**2 # cuandrado de los valores de x

# Este es un gráfico
plt.plot(x,y,'blue')
plt.title('Esto es un grafico')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
````

## Multigráficos 

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,10)
y = x**2 # cuandrado de los valores de x

# --- Forma 1 ---
# Si queremos hacer gráficos más pequeños, dividimos el espacio que tenemos para meter otros gráficos
# Grafico 1
plt.subplot(1,2,1) # numero de filas, numero de columnas, posicion
plt.plot(x,y,'blue')
plt.title('Grafico 1')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

# Grafico 2
plt.subplot(1,2,2) # numero de filas, numero de columnas, posicion
plt.plot(y,x,'green')
plt.title('Grafico 2')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

# Lo mostramos 
plt.show()

# --- Forma 2 ---
# Ahora vamos a crear 4 gráficos
figura, graficos = plt.subplots(nrows=2,ncols=2)

graficos[0][0].set_title('orange')
graficos[0][1].set_title('green')
graficos[1][0].set_title('red')
graficos[1][1].set_title('blue')

graficos[0][0].plot(x,y,'orange')
graficos[0][1].plot(x,y,'green')
graficos[1][0].plot(x,y,'red')
graficos[1][1].plot(x,y,'blue')

````

````python
# Construir 2 gráficos superpuestos, uno encima de otro, dentro de la misma figura
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,10)
y = x**2 # cuandrado de los valores de x

figura = plt.figure() # crea una figura, la cual podemos modificar sus propiedades, por ejem: add_axex

grafico1 = figura.add_axes([0.1,0.1,1,1]) # tamaño margen x, tamaño margen y, tamaño en x, tamaño en y
grafico1.plot(x,y,'orange')

grafico2 = figura.add_axes([0.2,0.5,0.3,0.5]) # tamaño margen x, tamaño margen y, tamaño en x, tamaño en y
grafico2.plot(x,y,'green')
````

## Tamaño de un gráfico y propiedades de la linea

````python
# Construir 2 gráficos superpuestos, uno encima de otro, dentro de la misma figura
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,10)
y = x**2 # cuandrado de los valores de x

figura = plt.figure(figsize=(10,4)) # crea una figura, la cual podemos modificar sus propiedades
grafico = figura.add_axes([0.1,0.1,1,1]) # tamaño margen x, tamaño margen y, % del tamaño en x (sobre 1), % del tamaño en y (sobre 1)

# metemos nuevas propiedades: color, ancho de linea, transparencia, estilo de la linea, puntos donde se unen las cordenadas
grafico.plot(x,y,color='orange',linewidth=3, alpha=0.5, linestyle='--',marker='o', markersize=10)

# Vamos a representar los valores X e Y en puntos en la gráfica
print("x: \n",x)
print("y: \n",y)
````

## Crear 2 gráficos juntos

````python
# Construir 2 gráficos en una
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,10)

figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,1,1]) # tamaño margen x, tamaño margen y, % del tamaño en x (sobre 1), % del tamaño en y (sobre 1)

grafico.plot(x,x**2,'orange',label='Cuadrado')
grafico.plot(x,x**3,'red',label='Cubo')
grafico.legend(loc=0) # loc=0 es la situación de la leyenda (situación óptima)
````
