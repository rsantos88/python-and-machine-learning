# Random Forest Tutorial: Bosques aleatorios

Random forest es una combinación de árboles de decisión, donde cada árbol selecciona una clase y luego se combinan decisiones de cada árbol, para seleccionar una clase final ganadora.
Es uno de los algoritmos de aprendizaje de clasificación con **mayor precisión**.
Funciona eficientemente en bases de datos grandes.
Puede manejar cientos de variables de entrada.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vinos = pd.read_csv('./datasets/vinos/original.csv')
```

```python
vinos
```

```python
# separamos entre características y columna objetivo
X = vinos.drop('Wine Type', axis=1)
y = vinos['Wine Type']
```

```python
# separamos los datos de test y entrenamiento
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
# entrenamos el modelo
from sklearn.ensemble import RandomForestClassifier
randomforest = RandomForestClassifier(n_estimators=80) # le pasamos el numero de estimadores (arboles uqe va a hacer para estimar el resultado)
randomforest.fit(X_train, y_train)
```

```python
predicciones = randomforest.predict(X_test)
```

```python
# si lo comparamos nos fijamos que lo ha predecido todo
dicc = {'Wine Type Test': y_test, 'Wine Type Predictes': predicciones}
pd.DataFrame(dicc)
```

```python
# la precisión es máxima en todos los vinos
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predicciones))
```

```python
# matriz de confusión
print(confusion_matrix(y_test, predicciones))
```
